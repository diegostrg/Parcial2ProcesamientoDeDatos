## Observación de la Variable `y`

La variable **`y`** es la **variable dependiente** o **variable objetivo** del dataset, ya que representa el resultado final que se desea predecir a partir del comportamiento de las demás variables.

- **Tipo de dato:** Categórico binario
- **Valores posibles:** `yes` / `no`
- **Rol:** Variable dependiente (Target)
- **Interpretación:** Indica si el cliente **aceptó** (`yes`) o **no aceptó** (`no`) el producto o servicio ofrecido en la campaña de marketing.

- Es importante revisar si existe **desbalance de clases** (muchos `no` y pocos `yes`), ya que esto puede sesgar el modelo.
- Debe mantenerse limpia, sin valores nulos ni categorías inconsistentes, porque cualquier error afecta directamente el entrenamiento del modelo.

```python
# ==========================================================
# Balanceo de la variable objetivo (y)
# ==========================================================
# Objetivo:
# Analizar la distribución de la variable dependiente "y"
# para identificar la proporción de cada clase (yes / no)
# y verificar si existe desbalance en el dataset.
#
# Importancia:
# Un conjunto de datos desbalanceado puede sesgar el
# entrenamiento de modelos de clasificación hacia la
# clase mayoritaria.
# ==========================================================


# ----------------------------------------------------------
# Cantidad total de registros
# ----------------------------------------------------------

cantTotal = df00.count()

print(f"Cantidad total de registros del dataset = {cantTotal}\n")


# ----------------------------------------------------------
# Conteo por categoría de la variable objetivo


cantY = df00.groupBy("y").count()


# ----------------------------------------------------------
# Cálculo del porcentaje de participación
# ----------------------------------------------------------

dfBalanceY = cantY.withColumn(
    "Porcentaje",
    (cantY["count"] * 100 / cantTotal)
)

# ----------------------------------------------------------
# Conversión a Pandas para visualización tabular
# ----------------------------------------------------------

tablaBalanceY = (
    dfBalanceY
    .toPandas()
    .sort_values("count", ascending=False)
)

print("Distribución de la variable objetivo:")
tablaBalanceY
```

```python
# Convertir a pandas
tablaBalanceY = (
    dfBalanceY
    .toPandas()
    .sort_values("count", ascending=False)
)

# Crear figura
plt.figure(figsize=(10,6))

# Barras
bars = plt.bar(
    tablaBalanceY["y"],
    tablaBalanceY["count"]
)

# Título y etiquetas
plt.title(
    "Distribución de la Variable Objetivo (Y)",
    fontsize=18,
    fontweight="bold",
    pad=20
)

plt.xlabel("Clase")
plt.ylabel("Cantidad de registros")

# Cuadrícula suave
plt.grid(axis="y", alpha=0.25)

# Etiquetas sobre barras
maximo = tablaBalanceY["count"].max()

for i, row in tablaBalanceY.iterrows():
    plt.text(
        i,
        row["count"] + maximo * 0.02,
        f'{row["count"]:,}\n({row["Porcentaje"]:.2f}%)',
        ha='center',
        fontsize=12,
        fontweight='bold'
    )

# Mostrar
plt.tight_layout()
plt.show()
```

## Observaciones del Gráfico de la Variable Objetivo `Y`

1. **Existe un desbalance marcado entre clases**  
   La categoría **`no`** concentra aproximadamente el **88.30%** de los registros, mientras que la categoría **`yes`** representa solo el **11.70%**. Esto evidencia una distribución claramente desigual.

2. **La clase mayoritaria puede sesgar el modelo**  
   Debido a la alta predominancia de la clase **`no`**, un modelo de clasificación podría inclinarse a predecir mayormente esta categoría, reduciendo su capacidad para identificar correctamente los casos positivos (`yes`).

3. **La clase positiva sigue teniendo representatividad analítica**  
   Aunque **`yes`** es la clase minoritaria, cuenta con una cantidad significativa de registros, lo que permite realizar análisis estadísticos y construir modelos predictivos sin que la clase positiva sea insuficiente para el aprendizaje.

4. **La distribución de `Y` sugiere un comportamiento de respuesta no uniforme**  
   La diferencia entre ambas categorías indica que la variable objetivo no presenta una distribución equilibrada, por lo que su comportamiento debe analizarse en relación con las variables predictoras para identificar qué factores influyen en la respuesta positiva.

```python
# ==========================================================
# Preparación de variables numéricas para correlación
# ==========================================================
# Objetivo:
# Construir un DataFrame únicamente con variables
# numéricas y transformar la variable objetivo "y"
# a formato binario (1/0).
#
# Importancia:
# La matriz de correlación de Pearson únicamente
# trabaja correctamente con variables numéricas.
# ==========================================================

from pyspark.sql import functions as F


# ----------------------------------------------------------
# Variables numéricas
# ----------------------------------------------------------

colNum = [
    'age',
    'balance',
    'day',
    'duration',
    'campaign',
    'pdays',
    'previous'
]


# ----------------------------------------------------------
# Construcción del DataFrame numérico
# ----------------------------------------------------------
# Se agregará también la variable objetivo "y"
# para analizar su relación con las demás variables.
# ----------------------------------------------------------

dfauxNum = df01.select(colNum + ['y'])


# ----------------------------------------------------------
# Transformación de la variable objetivo
# ----------------------------------------------------------
# Conversión:
#   yes -> 1
#   no  -> 0
#
# Esto permite:
#   - usar correlación de Pearson
#   - realizar análisis numérico
#   - evaluar relación entre variables y la variable objetivo
# ----------------------------------------------------------

dfauxNum = dfauxNum.withColumn(
    'y',
    F.when(dfauxNum['y'] == 'yes', 1)
     .otherwise(0)
)


# ----------------------------------------------------------
# Verificación del resultado
# ----------------------------------------------------------

dfauxNum.show(5)
```

```python
# ==========================================================
# Matriz de correlación de variables numéricas
# ==========================================================
# Objetivo:
# Analizar la relación lineal entre las variables
# numéricas y la variable objetivo "y".
#
# La correlación de Pearson permite identificar:
#   - relaciones positivas
#   - relaciones negativas
#   - dependencia entre variables
#   - variables con posible influencia sobre "y"
# ==========================================================

# ----------------------------------------------------------
# Ensamblar variables numéricas
# ----------------------------------------------------------
# VectorAssembler:
#   combina todas las columnas numéricas
#   en una sola columna vectorial llamada "features"
# ----------------------------------------------------------

ensamblador = VectorAssembler(
    inputCols=dfauxNum.columns,
    outputCol="features"
)

dfNumVector = (
    ensamblador.transform(dfauxNum).select("features")
)
# ----------------------------------------------------------
# Calcular matriz de correlación de Pearson
# ----------------------------------------------------------
matrizcorr = Correlation.corr(dfNumVector,"features","pearson").head()
# ----------------------------------------------------------
# Convertir matriz Spark a arreglo NumPy
# ----------------------------------------------------------

vectorcorr = matrizcorr[0].toArray()
# ----------------------------------------------------------
# Crear DataFrame de correlación
# ----------------------------------------------------------

dfCorrelacion = pd.DataFrame(
    vectorcorr,
    columns=dfauxNum.columns,
    index=dfauxNum.columns
)


# ----------------------------------------------------------
# Configuración visual del mapa de calor
# ----------------------------------------------------------

plt.figure(figsize=(10,7))

cmapJOHN = sns.diverging_palette(
    230,
    20,
    as_cmap=True
)
# ----------------------------------------------------------
# Graficar matriz de correlación
# ----------------------------------------------------------

sns.heatmap(
    dfCorrelacion,
    annot=True,
    fmt=".2f",
    cmap=cmapJOHN,
    linewidths=0.5
)

plt.title(
    "Matriz de Correlación de Variables Numéricas",
    fontsize=16,
    fontweight="bold"
)

plt.show()
```

```python
# ==========================================================
# Gráfica de distribución de la variable binaria "default"
# ==========================================================
# Objetivo:
# Visualizar la proporción de clientes que presentan
# crédito en mora (default) y aquellos que no.
#
# La gráfica de pastel permite identificar:
#   - distribución porcentual
#   - clase predominante
#   - balance de la variable binaria
# ==========================================================
cantBin = (
    df01..
        
        groupBy("default").count().toPandas()
)
# Separar ligeramente la primera porción
explode = (0.1, 0)

# Colores personalizados
colors = ['lightgreen', 'yellow']

plt.figure(figsize=(6,5))

cantBin.set_index("default")["count"].plot(
    kind='pie',explode=explode,colors=colors,autopct='%1.1f%%',startangle=90
)
plt.title(
    'Distribución de la Variable Binaria: default',
    fontsize=14,
    fontweight='bold'
)

# Ocultar etiqueta del eje Y
plt.ylabel('')


# ----------------------------------------------------------
# Mostrar gráfica
# ----------------------------------------------------------

plt.show()
```

```python
# ==========================================================
# Distribución de variables categóricas multidimensión
# ==========================================================
# Objetivo:
# Analizar la frecuencia de categorías presentes en
# variables categóricas con múltiples clases.
#
# Variables analizadas:
#   - job
#   - marital
#   - education
#   - contact
#   - month
#   - poutcome
#
# La gráfica de barras permite identificar:
#   - categorías dominantes
#   - distribución de frecuencias
#   - posibles desbalances
# ==========================================================

import matplotlib.pyplot as plt

variablesMulti = [
    'job',
    'marital',
    'education',
    'contact',
    'month',
    'poutcome'
]

for variable in variablesMulti:

    # Conteo de categorías
    cantMulti = (
        df01
        .groupBy(variable)
        .count()
        .toPandas()
    )

    # Orden descendente por frecuencia
    cantMulti = cantMulti.sort_values(
        'count',
        ascending=False
    )


    plt.figure(figsize=(5,2))

    plt.bar(
        cantMulti[variable],
        cantMulti['count']
    )

    plt.xlabel(variable.capitalize())

    plt.ylabel('Cantidad')

    plt.title(
        f'Distribución de la Variable: {variable}',
        fontsize=14,
        fontweight='bold'
    )

    plt.xticks(rotation=45)

    plt.grid(axis='y', alpha=0.25)

    plt.show()
```
